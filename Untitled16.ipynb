```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;
df=pd.read_csv("/content/drive/MyDrive/heart.csv")
df.head()
```

```python
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
print(df.target)
```

```python
sns.countplot(x=df.target,data=df,hue='cp',palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Disese','Disease'])
plt.title("Target Distribution");
```

```python
df.cp.value_counts()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend(["Haven't Disease","Have Disease"])
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,label='Haven\'t Disease',color='#1CA53B')
sns.distplot(df.age[df['target']==1],bins=30,label='Have Disease',color='red')
plt.legend()
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```
